```python
# superstore_analysis.ipynb

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset
df = pd.read_csv('SampleSuperstore.csv')

# 2. Quick look at the data
print(df.head())
print(df.info())
print(df.describe())

# 3. Check for duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
if duplicates > 0:
    df.drop_duplicates(inplace=True)
    print("Duplicates dropped.")

# 4. Check for missing values
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)

# (If necessary) Drop or impute missing values
# Example: df = df.dropna(subset=['Sales'])

# 5. Basic statistical analysis
total_sales = df['Sales'].sum()
print(f"Total Sales: {total_sales}")
total_profit = df['Profit'].sum()
print(f"Total Profit: {total_profit}")

# 6. Exploratory analysis: grouping and aggregations
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()
print(sales_by_category)

profit_by_region = df.groupby('Region')['Profit'].sum().reset_index()
print(profit_by_region)

# 7. Data visualization in Python
plt.figure(figsize=(8, 6))
sns.barplot(data=sales_by_category, x='Category', y='Sales')
plt.title('Total Sales by Category')
plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(data=profit_by_region, x='Region', y='Profit')
plt.title('Total Profit by Region')
plt.show()

# 8. Export cleaned data for Excel and Tableau usage
df.to_csv('cleaned_superstore.csv', index=False)
```
